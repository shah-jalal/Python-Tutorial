```python
import boto3
from datetime import datetime
```

```python
today = datetime.today()
```

```python
today
```

```python
todays_date = today.strftime("%Y%m%d")
```

```python
todays_date
```

```python
s3_client = boto3.client('s3')
```

```python
bucket_name = "luxxy-covid-testing-system-pdf-en-dj23"
```

```python
list_object_response = s3_client.list_objects(Bucket=bucket_name)
```

```python
list_object_response
```

```python
get_contents = list_object_response.get("Contents")
```

```python
get_contents
```

```python
get_all_s3_object = []

for item in get_contents:
    s3_object_name = item.get("Key")
    get_all_s3_object.append(s3_object_name)
```
